```python
# Author : Dony Varghese
# Date : 04-10-2022
# Synopsis : Given the numbers A, B and C, as well as the intervals in which three trucks are parked, 
# determine parking fee.

def get_input():
    a, b, c = input("Input three integers A, B and C (1≤C≤B≤A):").split(" ")
    s1, e1 = input("Input arrival and departure times for truck 1:").split(" ")
    s2, e2 = input("Input arrival and departure times for truck 2:").split(" ")
    s3, e3 = input("Input arrival and departure times for truck 3:").split(" ")

   # Convert the data types to integers and return the values properly.
    a = int(a)
    b = int(b)
    c = int(c)
    s1 = int(s1)
    e1 = int(e1)
    s2 = int(s2)
    e2 = int(e2)
    s3 = int(s3)
    e3 = int(e3)

   # Making have three tuples: t1, t2, t3 which are (starting minute, ending minute).

    # t1 is a tuple for truck 1
    # t2 is a tuple for truck 2
    # t3 is a tuple for truck 3
    # t1[0] is the start time for truck 1
    # t1[1] is the end time for truck 1
    
    t1 = (s1, e1)
    t2 = (s2, e2)
    t3 = (s3, e3)
    return a, b, c, t1, t2, t3

def calculate(a, b, c, t1, t2, t3):

    # We can create a time range using range()
    # built-in function. We need t1[1]+1 because
    # the end is exclusive, not inclusive.
    truck1 = range(t1[0], t1[1]+1)
    truck2 = range(t2[0], t2[1]+1)
    truck3 = range(t3[0], t3[1]+1)


    # We need to know the first starting point
    # which is the minimum of t1[0], t2[0], and t3[0]
    start_time = min(t1[0], t2[0], t3[0])
    # We also need to know the last ending point
    # which is the maximum of t1[1], t2[1], and t3[1]
    end_time = max(t1[1], t2[1], t3[1])

    # Now you need to loop through each time from
    # start_time to end_time and see if truck 1, 2, or 3
    # are there. Make a variable that keeps track of the
    # number of trucks.

    total_cost = 0
    for tm in range(start_time, end_time + 1):
        num_trucks = 0
        if tm in truck1:
            num_trucks += 1 #Keep track that truck 1 was here for this minute.

        if tm in truck2:
            num_trucks += 1 #Keep track that truck 2 was here for this minute.

        if tm in truck3:
            num_trucks += 1 #Keep track that truck 3 was here for this minute.
        if num_trucks == 1:
            total_cost = total_cost + a # Add a to the cost if there is only one truck
        elif num_trucks == 2:
            total_cost = total_cost + (2*b) # Add b to the cost for each truck if there are two trucks.
        else:
            total_cost = total_cost + (3*c)  # Add c to the cost for each truck if there are three trucks.

    return total_cost

if __name__ == "__main__":
    a, b, c, t1, t2, t3 = get_input()
    total_cost = calculate(a, b, c, t1, t2, t3)
    print(total_cost)
```
