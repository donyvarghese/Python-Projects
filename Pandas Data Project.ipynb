```python
# Author : Dony Varghese
# Date : 04-28-2022

# Synopsis :  Using data.gov as a data set repository to visualize 
# the COVID-19 vaccine distribution for the Moderna Pharmaceutical Corporation.

# Importing libraries

import pandas as pd



# Downloading the data set using Pandas

data = pd.read_csv("https://data.cdc.gov/api/views/b7pe-5nws/rows.csv?accessType=DOWNLOAD")

# Grouping by "Week of Allocations" and summing up the values for the same week. 
# This also removes the "Jurisdiction" column because it is not numbers.
s
data = data.groupby(by=["Week of Allocations"]).sum()


# Changing the "Week of Allocations" into a date for sorting purposes.

data.index = pd.to_datetime(data.index)



# up to the second week in March

data = data[(data.index<="2021-03-08")]


# Sorting "Week of Allocations" by date.

data = data.sort_index()

# Plotting the data by bar plot, size = 20x10.

ax = data.plot(kind='bar',width=.7,figsize=(20, 4),title='COVID-19 Vaccine Distribution for the Moderna Pharmaceutical Corporation', colormap='bwr')

# To remove the time from x axis. Changing the time format to index format.

ax = ax.set_xticklabels(data.index.format())





```
