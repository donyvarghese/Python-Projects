```python
#Importing required libraries
import math

loop=True #To keep repeating calculations until the user chooses quit.

while loop: #loop is False only if user enter 4. So, the programme is going to repeat until user input 4.
    
    # Asking for input
    print()
    print("1. Cartesian distance")
    print("2. Vector x matrix")
    print("3. Normalize")
    print("4. Quit")
    print ()
    user_input=int(input("Enter command: "))
    if user_input==1:
        values = input("Enter x1, y1, x2, and y2 seperated by spaces: ").split(" ")
        if len(values)!=4: #If user does not input 4 numbers.
            print ()
            print ("Enter 4 numbers!")
        else:
            values_list=[0]*4 #Defining a list of length 4 for 4 variables. If input is not given, it takes as zeroes.
            for i in range(len(values)):
                values_list[i] = float(values[i])
            
            dist=math.sqrt((values_list[2]-values_list[0])**2 + (values_list[3]-values_list[1])**2) # Distance equation sqrt((x2-x1)^2 + (y2-y1)^2)
            print ()
            print ("Cartesian distance is :", dist)
        
        loop=True #To keep while loop running
    elif user_input==2:
        values = input("Enter the vector coordinates v1, v2, v3 followed by 3x3 matrix in row-major order seperated by spaces: ").split(" ")
        if len(values)!=12: #If user does not input 12 numbers.
            print ()
            print ("Enter 12 numbers!")
        else:
            values_list=[0]*12 #Defining a list of length 12 for 12 variables. If input is not given, it takes as zeroes.
            for i in range(len(values)):
                values_list[i] = float(values[i])
            v=[0]*3  #Defining a list of length 3 for storing the values. 
            for i in range(3): #Loop for matrix mulitplication. To find first coordinate, wen need 0,1,2,3,4,5 elements of list. Then 0,1,2,6,7,8.
                v[i]= (values_list[0]*values_list[3*(i+1)]) + (values_list[1]*values_list[3*(i+1)+1]) + (values_list[2]*values_list[3*(i+1)+2])
                # This loop uses 0,1,2 always and as i changes, it changes 3,4,5 then 6,7,8 and 9,10,11.
            print ()
            print ("The vector after multiplication is:",v)
        loop=True #To keep while loop running
    elif user_input==3:
        
        values = input("Enter the vector coordinates v1, v2, and v3 seperated by spaces: ").split(" ")
        if len(values)!=3: #If user does not input 3 numbers.
            print ()
            print ("Enter 3 numbers!")
        else:
            values_list=[0]*3 #Defining a list of length 3 for 3 variables. If input is not given, it takes as zeroes.
            for i in range(len(values)):
                values_list[i] = float(values[i])
            sq_sum=0
            for i in range(3):
                sq_sum += (values_list[i]**2) # Taking sum of squares.
            length=math.sqrt(sq_sum) #Square root of sum of squares.
            norm_v=[0]*3 #Defining a list of length 3 for storing the values. 
            for i in range(3):
                norm_v[i]=values_list[i]/length #Dividing each term by length.
            print ()
            print ("The vector after normalization is:",norm_v)

        loop=True #To keep while loop running
    elif user_input==4:
        loop=False #To stop while loop
    else: #If user enter any other number.
        print ("The command you entered is wrong! Please enter correct command.")
        loop=True #To keep while loop running
        
        
        
```
